```python
import gensim
import snscrape.modules.twitter as sntwitter
import pandas as pd
import re
import nltk
import matplotlib.pyplot as plt
```

```python
tweets_nikol = pd.read_csv('nikol.csv')
tweets_nikol.head()
```

```python
preprocessed_tweets = []
for tweet in tweets_nikol['Tweets']:
    # remove URLs, hashtags, mentions
    tweet = re.sub(r'http\S+', '', tweet)
    tweet = re.sub(r'#\S+', '', tweet)
    tweet = re.sub(r'@\S+', '', tweet)
    # tokenize text
    tokens = nltk.word_tokenize(tweet)
    # remove stop words
    tokens = [word for word in tokens if word not in nltk.corpus.stopwords.words('english')]
    preprocessed_tweets.append(tokens)
```

```python
model = gensim.models.Word2Vec(preprocessed_tweets, vector_size=200, window=5, min_count=1, workers=4)
model.wv.save_word2vec_format('embedding.txt', binary=False)
```

```python
similarity = model.wv.similarity('Azerbaijan', 'attack')
print(similarity)
```

```python
similarity = model.wv.similarity('Armenia', 'attack')
print(similarity)
```

```python
similarity = model.wv.similarity('Armenia', 'democracy')
print(similarity)
```
